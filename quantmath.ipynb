### The three sacred rules  
A $\sigma$-algebra F on sample space $Ω$ must satisfy:  
**Rule 1** The whole space rule  
$Ω \in F$(the entire sample space must be in F)  
Why? we need to be able to say "something will definetly happen"   
**Rule 2:** The complement Rule  
If $A \in F$, then $A^c \in F$ (complements stay in F)  
Why? if we can measure "event A happens," we should also be able to measure "event A doesn't happen"  
**Rule 3:** The Union Rule  
If $\{A_1,A_2,A_3,\cdots\} \subset F$, then $ U_{i=1}^\infty A_i \in F$  
Why? if we can measure indiviidual events, we should be able to measure "at least one of these events happens"

# Conditional Probability and Independence
Focus: how to update probabilities when we get new information  

The basic formula: $P(A|B)=P(A\cap B)/ P(B) \neq 0$  
Read as the probability of A given that B has occured.  

To understand what it means think of it as updating your universe: 
Original universe $\Omega$  
New universe B(we know B happened)  
So the new question within this new universe B, what's the probability of A.

### Example: Fair coin Flipped Twice:
$\Omega = \{(H,H),(H,T),(T,H),(T,T)\}$ each equally likely (probability 1/4)  
Question: What's P(both heads | first flip is heads)?  
Step by step  
A = {(H,H)} = "both heads"  
B = {(H,H), (H,T)} = "first flip is heads"  
A∩B = {(H,H)} ∩ {(H,H), (H,T)} = {(H,H)}  

Calculate:  
P(A|B) = P(A∩B)/P(B) = P({(H,H)})/P({(H,H), (H,T)}) = (1/4)/(2/4) = 1/2  
Intuition: Once we know the first flip is heads, we're only considering {(H,H), (H,T)}. Within this reduced space, exactly half have both heads.  

**The mutliplication rule**
We can rewrite the equation to get P(A∩B) = P(A|B)P(B) = P(B|A)P(A)  

**The extended multiplication rule**
For multiple events:
P(A₁∩A₂∩...∩Aₙ) = P(A₁)P(A₂|A₁)P(A₃|A₁∩A₂)···P(Aₙ|A₁∩...∩Aₙ₋₁)  
event probabilities impact each other. say there are 3 outcomes u would use the multiplication rule. so the extended multiplication rule is essntially the multiplication rule repeated

**The Law of Total Probability: The "Divide and Conquer" Approach**  
Sometimes its easier to calculate P(A) by breaking it into pieces based on different scenarios.  
So say you have two disecting factors [even, $\leq 3$]  
Using Law of Total Probability:  
P(even) = P(even|B₁)P(B₁) + P(even|B₂)P(B₂)  
= (1/3)(1/2) + (2/3)(1/2) = 1/6 + 2/6 = 3/6 = 1/2   
Law of Total Probability says:  
"If you want P(A), but it's hard to calculate directly, then:  

Divide the world into mutually exclusive, exhaustive scenarios  
Calculate P(A|each scenario) - these are often easier!  
Weight each P(A|scenario) by P(scenario)  
Add them all up"

P(B|A) = P(A|B) × P(B) / P(A)

P(A|B) = Likelihood (how likely the evidence is, given your belief)  
P(A) = Total probability of seeing the evidence  
P(B|A) = Posterior belief (updated belief after seeing evidence)  
P(A) may be calculated using the law of total probability.

# Random Variable
a random variable is a random number that depends on the outcome of an experiment.  
A random variable X is a function X. $\Omega \rightarrow \mathbb{R}$ that assigns a real number to each outcome in the sample space.  
Requirement: For every real number x, the set $\{w \in \Omega : X(w) \leq x\}$ must be in our $\sigma$-algebra F.
This matters because  
The measurability condition
$\{w \in \Omega : X(w) \leq x\} \in F$ means:  
"the even that x is less than or equal to x must be observalble/measurable  
This ensures we can assign probabilities to statements like "$X \leq 50"  
In a finance situation $\rightarrow$  
Ω = all possible market scenarios  
X(ω) = stock price in scenario ω  
{X ≤ 50} = "stock price is at most $50" (this better be a measurable event!)  

**The cumulative distribution function(CDF)**
Defintion:  
F(x)=$F_x(x)=P(X \leq x)$
The cdf tells you the probability that the random variable takes a value less than or equal to x.
**Properties**
Non-decreasing: If x₁ < x₂, then F(x₁) ≤ F(x₂)  
Right-continuous: F(x) = lim_{y→x⁺} F(y)  
Boundary conditions:  lim_{x→-∞} F(x) = 0  
lim_{x→∞} F(x) = 1  



Why the cdf is non decreasing proof
Claim: If x₁ < x₂, then F(x₁) ≤ F(x₂)  
Proof:  
Let x₁ < x₂. We can write:  
{X ≤ x₂} = {X ≤ x₁} ∪ {x₁ < X ≤ x₂}  
Since these events are disjoint:  
P(X ≤ x₂) = P(X ≤ x₁) + P(x₁ < X ≤ x₂)  
Therefore:  
F(x₂) = F(x₁) + P(x₁ < X ≤ x₂)  
Since P(x₁ < X ≤ x₂) ≥ 0, we have:  
F(x₂) ≥ F(x₁) ✓

instead of density, discrete RVs have a mass function:

E[X]=$\sum_x x \cdot p(x)$  
Sum over all possible values x, weighted by their probabilities.  
Expected return = average return you'd expect over many repitions  
**Variance**
Var[x] = $E[(X-E[X])^2]= E[X^2]-(E[X])^2$  
How to calculate $E[X^2]$  
$E[X^2] = \sum_x x^2\cdot p(x)$

### Probability Density Function (PDF)
Since P(X = x) = 0 for continous RVs, we need a differrent approach: the density function.  
Definition: A continouous RV X has a proability density function (PDF) f(x) s.t $P(X \in B) = ∫_B f(x)dx$ for any interval B  
**Key properties of PDFs**  
1. f(x) $\geq$ 0 for all x (densities are non-negative)  
2. $∫_{-∞}^∞ f(x)dx = 1$ (total probability = 1)  
3. P(a < X $\leq$ b) = $∫_a^b f(x)dx$ (probability = area under curve)  

Going from PDF to CDF  
F(x) = P(X ≤ x) = ∫_{-∞}^x f(t)dt  
The CDF is the area under the PDF curve from -∞ to x.  

Going from CDF to PDF  
f(x) = dF(x)/dx (where F is differentiable)  
The PDF is the slope of the CDF.  

### Expected Value and Variance
E[X] = ∫_{-∞}^∞ x·f(x)dx  
For functions of X:  
E[g(X)] = ∫_{-∞}^∞ g(x)·f(x)dx

Var[X] = E[(X - E[X])²] = E[X²] - (E[X])² same as normal equations, but

# Stochastic Calculus
A stochastic process W = $(W_t)_{t\geq 0}$ if  
1. $W_0 = 0$  
2. Independent increments: $W_t-W_s$ is independent of the past  
3. Normal increments: $W_t - W_s$ ~ N(0,t-s) for t > s  
4. Continuous paths: The function t → Wₜ(ω) is continous  

E[Wₜ] = 0 for all t "on average, you end up where you started"  
Var[Wₜ] = t "uncertainty grows with time"  
Cov[Wₛ, Wₜ] = min(s,t) "earlier times are correlated with later times"  

what is $W_t$  
Think of Wₜ as: Cumulative random shocks to a stock price over time [0,t]  
W₁ - W₀: Random shock in first period  
W₂ - W₁: Random shock in second period  
Wₜ: Total accumulated randomness up to time t

this is an example of a switch/indicator function

## Itô Formula 
is the stochastic version of the chain rule. It tells us how to find the differential of a function of a stochastic process.  
Itô's Formula = Chain Rule + Stochastic Correction  
Itô's chain rule: if $Y_t = f(t, X_t)$, then you get an extra second derivative term.  
For an Itô process $dX_t = U_t dt + V_tdW_t$ and smooth function f(t,x):  
$$
dY_T = 
\left[
    \frac{\partial F}{\partial t}
    + \frac{\partial f}{\partial x} \cdot U_t
    + \frac{1}{2} \frac{\partial^2 f}{\partial x^2} \cdot v_t^2
\right] dt
+ \left[
    \frac{\partial f}{\partial x} \cdot V_t
\right] dW_t
$$
Uₜ = Drift coefficient (deterministic trend)  
Vₜ = Diffusion coefficient (volatility/randomness)  
In a stock Price Model dSₜ = μSₜdt + σSₜ dWₜ, then  
U = μSₜ (expected return rate)  
V = σSₜ (volatility × stock price) 



### For Brownian Motion(simplified Version)  
When $X_t = W_t$ (so U = 0, V = 1), itos formula becomes:  
$$
dY_t = 
\left[
    \frac{\partial{f}}{\partial{t}}
    + \frac{1}{2} \frac{\partial^2 f}{\partial x^2}
\right] dt
+ \left[
    \frac{\partial{f}}{\partial{t}}
\right] dW_t
$$
Breaking this down here is a step by tep recipie 
- Identify your function: f(t,x)  
- Caluculate derivatives: ∂f/∂t, ∂f/∂x, ∂²f/∂x²  
- Apply the formula: plug everything in  
- Simplify: collect dt and $dW_t$ terms  

### Example: d($W_t^2$)
**Step 1:** f(t,x) = $x^2$, so $Y_t = W_t^2$  
**Step 2:** Caluclate derivatives  
- $\frac{\partial{f}}{\partial{t}}=0$  
- $\frac{\partial{f}}{\partial{x}}=2x$ 
- $\partial^2{f}/\partial{x}^2=2$  
**step 3:** Apply Itô's formula  
d(W_t^2) = [0 + $\frac{1}{2} \cdot 2$] dt + [2Wₜ] dWₜ  
= dt + 2Wₜ dWₜ  
**Step 4:** Integrate to get  
$$ W_t^2 = \int_0^t ds + 2 \int_0^t W_s dW_s = t + 2\int_0^t W_s dW_s$$
Therefore  $$\int_0^t W_s dW_s = (W_t^2-t)/2$$

More examples  
Example 1: d(Wₜᵖ) for p ≥ 2  

f(x) = xᵖ  
∂f/∂x = pxᵖ⁻¹, ∂²f/∂x² = p(p-1)xᵖ⁻²  
Result: dWₜᵖ = pWₜᵖ⁻¹ dWₜ + [p(p-1)/2]Wₜᵖ⁻² dt  

Example 2: d(sin(Wₜ))  

f(x) = sin(x)  
∂f/∂x = cos(x), ∂²f/∂x² = -sin(x)  
Result: d(sin(Wₜ)) = cos(Wₜ) dWₜ - ½sin(Wₜ) dt  

### So when to use Itô's
You use it whenever you have:
- A function f(t, x)
- Applied to a stochastic process (like brownian motion or any Itô process)
- Any you want to find how that function evolves over time

Use this for pattern recognition and use itos formuala  
- f(Wₜ) = Wₜ², sin(Wₜ), eᵂᵗ, etc.
- f(t, Wₜ) = tWₜ², e^(t/2)sin(Wₜ), etc.
- f(Sₜ) where Sₜ follows geometric Brownian motion

# Black-Scholes PDE: The crown jewel of mathematical finance

Lets derive it  

**Step 1: The Setup**  

$$dS_t = \mu S_t dt + \sigma S_t dW_t$$

- $\mu$ = expected return (drift)
- $\sigma$ = volatility 
- $S_t$ = stock price at time t

Option value: $f(t, S_t)$ = some function of time and stock price  

---

**Step 1.5** I want to add where the parts of the equation come from so it makes more sense  

1. starting with $dS_t = \mu S_t\, dt + \sigma S_t\, dW_t$  
    - $\mu S_t\, dt$ → deterministic drift term (average rate of return)  
    - $\sigma S_t\, dW_t$ → random diffusion term (volatility × Brownian motion)

2. First term:  
$$\frac{\partial f}{\partial t} dt$$
This is the explicit time change in the option price.  

3. $\frac{\partial f}{\partial S} \cdot dS_t$

$$= \frac{\partial f}{\partial S} (\mu S_t dt + \sigma S_t dW_t)$$

This is the first-order effect of changes in $S_t$ on the option price.

4. $\frac{1}{2} \frac{\partial^2 f}{\partial S^2} (dS_t)^2$

Since $(dS_t)^2 = (\sigma S_t)^2 (dW_t)^2 = \sigma^2 S_t^2 dt$, this term becomes:

$$\frac{1}{2} \sigma^2 S_t^2 \frac{\partial^2 f}{\partial S^2} dt$$

This is the second-order correction from the curvature (convexity) of f in S — the part that has no analog in ordinary calculus. convexity adjustment from volatility; even if the mean change in $S_t$ is zero, volatility changes the expected option value.  

5. $\sigma S_t \frac{\partial f}{\partial S} \, dW_t$ → randomness in option price due to stock price fluctuations.

---

**Step 2:** Apply Itô's formula to the option. Since f depends on both t and $S_t$, we use the general Itô formula:  

$dX_t = \mu S_t dt + \sigma S_t dW_t$ (so $U = \mu S_t$, $V = \sigma S_t$)

$$
df =
\left[
    \frac{\partial f}{\partial t} +
    \frac{\partial f}{\partial S} \cdot \mu S_t + 
    \frac{1}{2}\frac{\partial^2 f}{\partial S^2} \cdot (\sigma S_t)^2
\right] dt +
\left[
    \frac{\partial f}{\partial S} \cdot
    \sigma S_t
\right] dW_t
$$

$$
df = 
\left[
    \frac{\partial f}{\partial t} +
    \mu S_t \frac{\partial f}{\partial S} +
    \frac{1}{2}\sigma^2 S_t^2 \frac{\partial^2 f}{\partial S^2}
\right] dt
+ 
\left[
    \sigma S_t \frac{\partial f}{\partial S}
\right] dW_t
$$

---

**Step 3: The delta hedging Strategy**  

Create a portfolio:  
- Long 1 option: Value = $f(t, S_t)$  
- Short $\Delta$ shares: Value = $-\Delta S_t$  
- Total portfolio value: $\Pi = f - \Delta S_t$  

Where $\Delta = \frac{\partial f}{\partial S}$ (this choice eliminates risk!)  

---

**Step 4: Portfolio Evolution**

Change in portfolio value:  
$$d\Pi = df - \Delta dS_t$$

so between the two steps we are going from the option price $f(t,S_t)$ to the portfolio above.  

We know $df$ from step 2 so substitute into our expression  
and we also know $S_t$ from the stock step: $dS_t = \mu S_t dt + \sigma S_t dW_t$

$$
d\Pi =
\underbrace{\left[\frac{\partial f}{\partial t} + \mu S_t \frac{\partial f}{\partial S} + \frac{1}{2}\sigma^2 S_t^2 \frac{\partial^2 f}{\partial S^2} \right] dt + \left[ \sigma S_t \frac{\partial f}{\partial S} \right] dW_t}_{\text{from } df}
\;-\;
\underbrace{\Delta\left[ \mu S_t dt + \sigma S_t dW_t \right]}_{\text{from } -\Delta dS_t}
$$

---

**Pause for a second and realize the importance of this step**

The reason for doing this subtraction. (eliminate risk)  

Goal: eliminate the $dW_t$ term (the random part) by choosing $\Delta = \frac{\partial f}{\partial S}$ → this is delta hedging.  

with $dW_t$ gone the portfolio becomes riskless for an instant and a riskless portfolio must grow at the risk-free rate $r$, not at $\mu$.  

---

**Simplify and see the magic**

$$d\Pi = \left[\frac{\partial f}{\partial t} + \frac{1}{2}\sigma^2 S_t^2 \frac{\partial^2 f}{\partial S^2}\right] dt$$

---

**Step 5: No-Arbitrage Argument**   

Since the portfolio has no randomness, it must earn the risk-free rate $r$:  

$$d\Pi = r\Pi \, dt = r(f - \Delta S_t) dt = r \left(f - S_t \frac{\partial f}{\partial S}\right) dt$$

---

**Step 6: Set Equal and Derive PDE**

$$\left[ 
    \frac{\partial f}{\partial t} + 
    \frac{1}{2}\sigma^2 S_t^2 \frac{\partial^2 f}{\partial S^2}
\right] dt =
r\left(f - S_t \frac{\partial f}{\partial S}\right) dt$$

Divide by $dt$:

$$\frac{\partial f}{\partial t} + 
    \frac{1}{2}\sigma^2 S_t^2 \frac{\partial^2 f}{\partial S^2} =
rf - rS_t \frac{\partial f}{\partial S}$$

Then push to one side and we have the Black-Scholes PDE:

$$\frac{\partial f}{\partial t} + 
    \frac{1}{2}\sigma^2 S_t^2 \frac{\partial^2 f}{\partial S^2} -
rf + rS_t \frac{\partial f}{\partial S} = 0$$

With boundary condition: $f(T, S_t) = \text{payoff function}$  

So the option price $f(t,S)$ must satisfy this relationship at every point in time $t$ before expiry and for every stock price $S$.  

---

**Breaking down $f(T, S_t) = \text{payoff function}$**   

So this is saying that  
- at the final time $T$ (option expiry),  
- there is no more uncertainty remains - the option's value is just whatever the payoff is for that stock price $S_t$  
- in finance PDEs the "boundary" in time is the expiry date.  
- in other words "When the option expires at time $T$, the price $f$ is exactly equal to the payoff of the option — this is the condition we use to solve the PDE backwards in time."

    # Solving the Black–Scholes Formula to get application and proof of Feynman–Kac theorem

    Take

    $$
    \frac{\partial f}{\partial t} +
    \frac{1}{2} \sigma_t^2 \frac{\partial^2 f}{\partial S^2}
    - r f + r S_t \frac{\partial f}{\partial S} = 0
    $$

    With boundary condition:

    $$
    f(T, S) = \Phi(S) \quad \text{(payoff function)}
    $$

    To solve the PDE we use **Feynman–Kac theorem**.

    ---

    **Step 1: Define the Candidate Solution**

    $$
    v(t, x) = \mathbb{E}\left[ e^{-r(T-t)} \, g(X_T) \mid X_t = x \right]
    $$

    *wtf is this?*

    $\rightarrow$ $v(t,x)$ can be thought of as a particle beginning at position $x$ at time $t$.  
    Let it evolve according to:

    $$
    dX_s = b\,ds + a\,dW_s
    $$

    (just a basic stochastic calculus problem)

    - At time $T$, the particle is at position $X_T$  
    - Apply function $g$ to get $g(X_T)$  
    - Discount it back:

    $$
    e^{-r(T-t)} g(X_T)
    $$

    ---

    **Side note**: what does *discount it back* mean?  

    The term:

    $$
    e^{-r(T-t)}
    $$

    is the discount factor for interest rate $r$, over the time span from $t$ to $T$.  
    In words: "If I'm promised $g(X_T)$ dollars at time $T$, what is that worth at time $t$ right now?"

    - Take the expectation of this: $v(t,x)$ is this expected discounted value.

    ---

    ### A change in time

    The math where $dt$ is tiny:

    $$
    v(t + dt, x) =
    \mathbb{E}\left[
    e^{-r(T - (t+dt))} g(X_T)
    \mid X_{t+dt} = x
    \right]
    $$

    $$
    = \mathbb{E}\left[
    e^{-r(T - t - dt)} g(X_T) \mid X_{t+dt} = x
    \right]
    $$

    $$
    = \mathbb{E}\left[
    e^{r\,dt} \cdot e^{-r(T - t)} g(X_T)
    \mid X_{t+dt} = x
    \right]
    $$

    Here $e^{r\,dt} \approx 1 + r\,dt$ for small $dt$, so there is an infinitesimal change.

    ---

    ### Expanding $v$ with Itô

    We want to apply Itô's formula to $v(t+dt, X_{t+dt})$.  
    Since $X_{t+dt}$ follows $dX = b\,dt + a\,dW$:

    $$
    dv =
    \frac{\partial v}{\partial t} \, dt
    + \frac{\partial v}{\partial x} \, dX
    + \frac{1}{2} \frac{\partial^2 v}{\partial x^2} (dX)^2
    $$

    $$
    = \frac{\partial v}{\partial t} \, dt
    + \frac{\partial v}{\partial x} (b\,dt + a\,dW)
    + \frac{1}{2} \frac{\partial^2 v}{\partial x^2} a^2 \, dt
    $$

    $$
    = \left[ \frac{\partial v}{\partial t}
    + b \frac{\partial v}{\partial x}
    + \frac{1}{2} a^2 \frac{\partial^2 v}{\partial x^2} \right] dt
    + a \frac{\partial v}{\partial x} \, dW
    $$
    ---
    Find the expectation
    $$
    E[dv]= 
    \left[ \frac{\partial v}{\partial t}
    + b \frac{\partial v}{\partial x}
    + \frac{1}{2} a^2 \frac{\partial^2 v}{\partial x^2} \right] dt
    $$
    The dW term has zero expectiation
    ---

    ### The martingale property

    From our expectation relationship:
    $$
    v(t,x) = E[e^{(-r·dt)} v(t+dt, X_{t+dt})]
    $$
    Now we want to expand with $e^{(-r\cdot dt)}\approx 1 - r \cdot$ dt:
    $$
    v((t,x)) = E[(1 - r \cdot dt)(v(t,x) + dv)]
    $$
    $$
    = E[v(t,x) + dv - r \cdot dt \cdot v(t,x) - r \cdot dt \cdot dv]
    $$
    since dt is small, ignore the r $\cdot$ dt dv term:  
    $$
    v(t,x) \approx E[v(t,x) + dv - r \cdot dt \cdot v(t,x)]  
    $$
    $$
    = v (t,x) + E[dv] - r \cdot dt \cdot v(t,x)
    $$
    $$
    v(t,x) - v(t,x) = v(t,x) + E[dv] - r·dt·v(t,x) - v(t,x)
    0 = E[dv] - r·dt·v(t,x)
    $$
    which simplifies to
    $$
    0 = E[dv] - r·dt·v(t,x)
    $$
    ---
    ### finale
    So we have 
    $$
    0 = \left[ \frac{\partial v}{\partial t}
    + b \frac{\partial v}{\partial x}
    + \frac{1}{2} a^2 \frac{\partial^2 v}{\partial x^2} \right] dt 
    - r\cdot dt \cdot v(t,x)
    $$
    Divide by dt 
    $$
    0 = \frac{\partial v}{\partial t}
    + b \frac{\partial v}{\partial x}
    + \frac{1}{2} a^2 \frac{\partial^2 v}{\partial x^2} 
    - r \cdot v(t,x)
    $$
    Rerange it to make it all pretty
    $$
    \frac{\partial v}{\partial t}
    + \frac{1}{2} a^2 \frac{\partial^2 v}{\partial x^2} 
    + b \frac{\partial v}{\partial x}
    - r \cdot v(t,x)
    $$
    This is our beautiful pde we want

    # Math behind Feynman-Kac
    Lets start with the PDE
    $$
    \frac{\partial f}{\partial t}
    + rS \frac{\partial f}{\partial S}
    + \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 f}{\partial S^2}
    - rf = 0
    $$
    Rewrite it to match the Feynman-Kac Pattern
    $$
    \frac{\partial f}{\partial t}
    + \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 f}{\partial S^2}
    + rS \frac{\partial f}{\partial S}
    - rf = 0
    $$
    - x = S (our variable is stock price)  
    - a(S) = $\sigma$ S (diffusion coefficent)  
    - b(S) = rS (drift coefficient)  
    - r = r (discount rate)  
    - g(s) = Φ(S) (payoff function)  

    $$
    dS_s = b(S_s) ds + a(S_s) dW_s
    $$
    $$
    dS_s = rS_s ds + σS_s dW_s
    $$
    This is the stock price under the risk-neutral measure  
    The original stock had drift μ, but under the risk-neutral measure, it has drift r.  

    ### Whats the solution formula?
    $$
    f(t,S) = E^Q[e^(-r(T-t)) Φ(S_T) | S_t = S]
    $$
    Means "The option value today equals the expected discounted payoff under the risk-neutral measure"  
    But we still need to calculate this expectation!  
    We need to solve $dS_s = rS_s ds + σS_s dW_s$  
    This is a geometric Brownian motion. Let's use Itô's formula on $Y_s = ln(S_s)$:  
    Let f(S) = ln(S), so:  
    ∂f/∂S = 1/S  
    ∂²f/∂S² = -1/S²  
    Apply itos fomula
    $$d(ln S_s) = ∂f/∂S · dS_s + ½ ∂²f/∂S² · (dS_s)²$$
    substituting $dS_s = rS_s ds + σS_s dW_s$
    $$d(ln S_s) = \frac{1}{S} \cdot (rS_s ds + σS_s dW_s)
    + \frac{1}{2} (\frac{-1}{S^2})(rS_s ds + σS_s dW_s)^2$$ 
    ---
    **Quick fukin stupid quadratic rules to make your life harder**
    Expand $(dS_s)^2$
    $$
    (dS_s)^2 = (rS_s ds + \sigma S_s dW_s)^2
    $$
    $$
    = (rS_s ds)^2 
    + 2(rs_s ds)(\sigma S_s dW_s)
    + (\sigma S_s dW_s)^2
    $$
    Now apply the multiplication rules  
    - $(ds)^2$ =0 (deterministic terms squared vanish)  
    - ds $\cdot dW_s =0$ (mixed terms vanish)  
    - $(dW_s)^2$ = ds(this is the key)  
    **SO**
    $$
    (dS_s)^2 = 
    0 + 0 + (\Sigma S_s)^2 ds = \Sigma^2 S_s^2 ds
    $$
    ---
    $$d(ln S_s) = (\frac{1}{S_s})\left[rS_s ds + σS_s dW_s\right]
    + \frac{1}{2} (\frac{-1}{S_s^2})(σS_s dW_s)^2$$ 
    Simplify
    $$
    d(ln S_s) = r ds + σ dW_s + \frac{1}{2}(\frac{-1}{S_s^2})(σ^2S_s^2 ds)
    $$
    $$
    d(ln S_s) = r ds + σ dW_s - \frac{1}{2}σ^2 ds
    $$
    $$
    d(ln S_s) = (r - \frac{1}{2}σ²) ds + σ dW_s
    $$
    Integrate the SDE
    From time t to T:
    $$
    ln(S_T) - ln(S_t) = (r - \frac{1}{2}\sigma^2)(T_t) + \sigma (W_T - W_t)
    $$
    Since $W_T - W_t ~ N(0,T-t)$, we can write $W_T - W_t$ as $\sqrt{(T-t)Z}$ where Z~N(0,1):
    $$
    ln(S_T) = ln(S_t) + (r - \frac{1}{2}\sigma^2)(T_t) + \sigma\sqrt{(T-t)Z}
    $$
    Taking Exponentials:
    $$
    S_T = S_t exp[(r-\frac{1}{2}\sigma^2)(T-t) + \sigma\sqrt{(T-t) Z}]
    $$
    This is the distribution of $S_T$ under the risk-neutral meausure

    ---
    **Calculating the expectation**
    For a european call with payoff $\phi(S_T) = max(S_t - k, 0):
    $$
    C(t,S_T) = e^{(-r(T-t)) E^Q[max(S_T - k, 0)]}
    $$
    Substituing our expression for $S_T$:
    $$
    C(t,S_t) = e^{(-r(T-t))} E^Q[max(S_t exp[(r - \frac{1}{2}\sigma^2)(T-t) + \sigma\sqrt{(T-t) Z}] - K, 0)]
    $$
    Where Z ~ N(0,1)  
    This is where heavy normal dustribution calculations come in 
    The option pays off when
    $$
    S_t exp[(r - ½σ²)(T-t) + σ√(T-t) Z] > K
    $$
    ---
    Ok so we have to split the expectation. idek what that really means, but the max function is awkward to integrate so we use this trick.
    $$
    max(S_T - K, 0) = \begin{cases}
        S_T - K, & \text{if } S_T > K \\
        0,  & \text{if } S_T \leq K
    \end{cases}
    $$
    We can rewrite this using an indicator function
    $$
    max(S_T - K, 0) = (S_T - K) \cdot \mathbb{I}_{S_T > K}
    $$
    Where $\mathbb{I}_{S_T > K}$ is the indicator function:
    $$
    \mathbb{I}_{S_T > K} = \begin{cases}
        1, & \text{if } S_T > K \\
        0,  & \text{if } S_T \leq K
    \end{cases}
    $$
    So our expectation becomes:
    $$
    E[max(S_T - K, 0)] = E[(S_T - K) · \mathbb{I}_{S_T > K}]
    $$
    $$
    = E[S_T · \mathbb{I}_{S_T > K}] - K · E[\mathbb{I}_{S_T > K}]
    $$
    ---
    What are these two integrals?  
    **Integral 1** $E[\mathbb{I}_{S_T > K}]$  
    This is P$(S_T>K)$ - the probability that the option finishes in the money  
    **Integral 2:** $E[S_T · \mathbb{I}_{S_T > K}]$   
    This is the expected value of $S_T$, given that $S_T > K$  

    ---
    We know that $S_T = S_t e^[(r - \frac{1}{2}\sigma^2)(T-t) + \sigma\sqrt{(T-t) Z}]$ so $S_T>K$ when $$S_t e^[(r - \frac{1}{2}\sigma^2)(T-t) + \sigma\sqrt{(T-t) Z}] > K$$ 
    **taking logs**
    $$
    ln(S_t) (r - \frac{1}{2}\sigma^2)(T-t) + \sigma\sqrt{(T-t) Z} > ln(K)
    $$
    $$
    \sigma\sqrt{(T-t) Z} > \frac{ln(K)}{S_t} - (r - \frac{1}{2}\sigma^2)(T-t)
    $$
    $$
    Z > [ln (\frac{K}{S_t}) - (r - \frac{1}{2}\sigma^2)(T-t)]
    $$
    Define this as
    $$
    d_2 = [ln (\frac{K}{S_t}) - (r - \frac{1}{2}\sigma^2)(T-t)]
    $$
    So $S_T >K$ when z > $-d_2$

    ---
    ### Calculate integrals
    1.
    $$
    E[\mathbb{I}_{S_T > K}] = P(S_T > K) = P(Z > d_2) = P(Z < d_2) = N(d_2)
    $$
    2.
    $$
    E[S_T \cdot \mathbb{I}_{S_T > K}] = E[S_t exp[(r-\frac{1}{2}\sigma^2)(T-t) + \sigma\sqrt{(t-t)} Z] \cdot \mathbb{I}_{Z > -d_2}]
    $$
    This is where we need the change of measure trick. Let me explain this step by step.  

    ---
    ### The change of measure technique
    The problem: we have Z ~ N(0,1) in the integral, but the exponential term makes it hard to integrate.  
    The solution: change to a new probability measure where the math is easier!  

    $$
    d_1 = d_2 + \sigma\sqrt{(T-t)} = \frac{[ln(\frac{S_t}{k}) + (r + \frac{1}{2}\sigma^2)(T-t)]}{\sigma\sqrt{(T-t)}}
    $$

    After some measure theory magic(which involves changing from measure Q to a new measure), this becomes:  
    $$
    E^Q[S_T \cdot \mathbb{I}_{S_T > K}] = S_t e^(r(T-t)) N(d_1)
    $$
    ---
    $$
    C(t,S_t) = e^{(-r(T-t))} [S_t e^{(r(T-t))} N(d_1) - K N(d_2)]
    $$
    $$
    = e^{(-r(T-t))} S_t e^{(r(T-t)) N(d_1)} - e^{(-r(T-t)) K N(d_2)}
    $$
    $$
    = S_t N(d_1) - Ke^{(-r(T_t))} N(d_2)
    $$
    where 
    $d₁ = [ln(S/K) + (r + ½σ²)(T-t)] / [σ√(T-t)] \\
    d₂ = [ln(S/K) + (r - ½σ²)(T-t)] / [σ√(T-t)] = d₁ - σ√(T-t)$






```python
# Monte Carlo pricing of a European call using risk-neutral GBM, plus closed-form Black-Scholes check.
# This code will run here and print both Monte Carlo estimate and exact Black-Scholes price.

import numpy as np
from math import log, sqrt, exp, erf

def normal_cdf(x):
    """Standard normal CDF via error function (no external libs)."""
    return 0.5 * (1 + erf(x / sqrt(2)))

def bs_call_price(S0, K, r, sigma, T):
    """Black-Scholes closed-form price for European call (no dividends)."""
    if T <= 0:
        return max(S0 - K, 0.0)
    d1 = (log(S0 / K) + (r + 0.5 * sigma**2) * T) / (sigma * sqrt(T))
    d2 = d1 - sigma * sqrt(T)
    return S0 * normal_cdf(d1) - K * exp(-r * T) * normal_cdf(d2)

def mc_call_price(S0, K, r, sigma, T, n_paths=200000, seed=42):
    """Monte Carlo estimate of discounted expected payoff for a European call."""
    rng = np.random.default_rng(seed)
    # Use analytic terminal distribution of GBM to sample S_T directly
    mu = (r - 0.5 * sigma**2) * T
    vol = sigma * sqrt(T)
    Z = rng.standard_normal(n_paths)
    ST = S0 * np.exp(mu + vol * Z)
    payoffs = np.maximum(ST - K, 0.0)
    discounted = np.exp(-r * T) * payoffs
    price = discounted.mean()
    stderr = discounted.std(ddof=1) / np.sqrt(n_paths)
    return price, stderr

# Example parameters (you can edit these)
S0 = 100.0   # spot price
K  = 100.0   # strike
r  = 0.05    # risk-free rate (annual)
sigma = 0.2  # volatility (annual)
T = 1.0      # time to maturity (years)

mc_price, mc_stderr = mc_call_price(S0, K, r, sigma, T, n_paths=200000)
bs_price = bs_call_price(S0, K, r, sigma, T)

print(f"Parameters: S0={S0}, K={K}, r={r}, sigma={sigma}, T={T}")
print(f"Monte Carlo price (n=200k): {mc_price:.6f}  ± {1.96*mc_stderr:.6f} (95% CI)")
print(f"Black-Scholes exact:        {bs_price:.6f}")
print(f"Difference (MC - BS):       {mc_price - bs_price:.6e}")
```
